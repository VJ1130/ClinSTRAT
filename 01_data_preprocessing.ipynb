```python
import pandas as pd
print("Pandas version:", pd.__version__)
```

```python
df = pd.read_csv('brca_metabric/data_clinical_patient.txt', sep='\t', comment='#')
#sep='\t' specifies that the file is tab-separated, and comment='#' indicates that lines starting with '#' should be treated as comments and ignored.
clinical_patient = df
display(clinical_patient.head(n=10))
```

```python
# the following line will display the summary information of the clinical_patient DataFrame, including the number of non-null entries, data types, and memory usage.
clinical_patient.info()
```

```python
# Subset the dataframe
#.copy() is used to create a new DataFrame that is a copy of the selected columns from the original DataFrame. This is important to avoid SettingWithCopyWarning in pandas, which can occur when trying to modify a slice of a DataFrame.
clinical_baseline = clinical_patient.copy()
display(clinical_baseline.head())
```

```python
# 1. Define the target variable (y) from the clean clinical_patient DataFrame.
y = clinical_baseline['CLAUDIN_SUBTYPE']

# 2. Define the feature matrix (X) by dropping the columns that are not needed for the model.
cols_to_drop = [
    'PATIENT_ID',
    'CLAUDIN_SUBTYPE',  # Target variable
    'THREEGENE', 'COHORT', 'HER2_SNP6',  # Target leakage
    'OS_MONTHS', 'OS_STATUS',
    'RFS_MONTHS', 'RFS_STATUS', 'VITAL_STATUS',  # Endpoints / Validation
    'SEX'  # Uninformative constant
]

# 3. Create feature matrix (X) by dropping the specified columns from the clean clinical_patient DataFrame.
X = clinical_baseline.drop(columns=cols_to_drop, inplace=False) #inplace=False ensures that the original DataFrame is not modified, and a new DataFrame is returned instead.
display(X.head(n=10))
```

```python
print("Shape of feature matrix (X):", X.shape)

print("Shape of target variable (y):", y.shape)
print("Target class counts:\n", y.value_counts())
print(y.value_counts(normalize=True))  # Display class distribution as percentages
print(y.value_counts(dropna=False))  # Display class distribution including NaN values
display(X.head(n=10))
```

```python
# Check for missing values in the feature matrix (X)
missing_values = X.isnull().sum()
print("Missing values in each column:\n", missing_values)
```
